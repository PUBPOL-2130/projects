### Literature Review

Racial disparities in a multitude of areas in the United States have been the focus of considerable study in the past 50 years. Racial disparities were first investigated in health contexts, but the area of research has seen manifestations in many areas such as housing, health, education, and, more recently, the criminal justice system. An understanding of this landscape begins with an analysis of the systemic structures that have historically and continuously disadvantaged minority groups across the country. These disparities are not only social and economic but are codified in policy, institutional practice, and cultural attitudes regarding race. The 1985 Report of the Secretary's Task Force on Black and Minority Health (U.S. Department of Health and Human Services) identified persistent health disparities as reflections of more basic societal inequalities, many of which overlap with the criminal justice realm. Scholars such as Gibbons (2005) then traced such systemic disparities to general societal structures, including unequal access to resources and institutional discrimination.

Within the broader subject area of racial inequality, the criminal justice system has been one of the most studied areas. Scholars and activists have been amply documented in demonstrating how police procedures, prosecutorial discretion, sentencing laws, and parole decisions all disproportionately harm communities of color. According to The Sentencing Project's report Mass Incarceration Trends (2024), the United States incarcerates more people than any other country in the world, and the racial disproportion in these statistics is staggering. Black Americans are incarcerated at more than five times the rate of white Americans (One in Five, Barry & Celeste, 2023), and this inequality cannot be accounted for by differences in crime rates alone.

As national public interest in the carceral system has increased, academic scholarship has increasingly focused on racial disparities in incarceration. In his groundbreaking work, Punishment and Inequality in America (2006), Bruce Western points out how mass incarceration reflects as well as intensifies racial and class inequality. Western shows how incarceration decreases economic prospects, destabilizes families, and compromises civic engagement—effects that are particularly acute in Black communities. This is compounded by intergenerational trends: a report by the Brookings Institution titled "Work and opportunity before and after incarceration" quotes that individuals from lower-income families are much more likely to be incarcerated, and that racial disparities are inextricably linked with class-based disparity.

Turning to Louisiana, the state has consistently had one of the highest incarceration rates in the country. The Vera Institute's report "Louisiana Locked Up: A Problem in Every Parish" (n.d.) determines that nearly every parish has pretrial incarceration rates far exceeding national norms. Furthermore, the Prison Policy Initiative's "Mass Incarceration: The Whole Pie 2025" supplements that Louisiana's carceral system is particularly reliant on local jails, which often hold people for extended periods without trial. These patterns have precipitated special attention to Louisiana's racial dynamics. The Louisiana Law Review article "How and Why Race Continues to Influence the Administration of Criminal Justice in Louisiana" presents a legal and historical analysis of the ways in which racial bias is so embedded that it impacts arrest, charging, and sentencing determinations. To augment this, the Southern Poverty Law Center (2018) details the extent of racial profiling in Louisiana policing, arguing that such practices are both unconstitutional and ineffective.

There is a vast literature attempting to explain why racial groups incur differing rates of incarceration. One of the most favored explanations is differential policing of communities, resulting in higher arrest rates in Black communities orthogonal to actual crime rates. Prosecutorial discretion and mandatory minimum sentencing, which remove the context and nuance from judicial decisions, are another explanation. Employment and economic opportunity are also essential. As the Brookings Institution outlines, former offenders face significant barriers in the employment market, magnifying the cumulative effects of poverty, which, in the U.S., disproportionately fall on Black families. The "The poorer your parents are, the more likely you are to be incarcerated" data visualization graphically displays the intersectionality of class and risk of incarceration.

In Louisiana, these national trends are acted out with special intensity due to the state's unique criminal justice policies and local government structures. Parishes exert a lot of influence over the operation of jails, and the state has historically depended on incarceration as a solution to a range of social ills. Finally, the intersection of poverty, lack of legal resources, and punitive local politics intensifies the racial disparities that the data identify.

Lastly, it is necessary to view the overall national policy environment to comprehend the development of these disparities between 1990 and 2010. The 1994 Violent Crime Control and Law Enforcement Act, which is also referred to as the federal crime bill, increased prison funding and promoted the utilization of tougher sentencing, such as "three strikes" laws and truth-in-sentencing laws. Based on the Brennan Center's "What Caused the Crime Decline? These policies did not reduce crime significantly but assisted in growing the prison population. Louisiana included numerous provisions at the state level, which further solidified punitive responses to crime.

The Sentencing Project highlights that while there have been some reforms in recent years aimed at reducing disparities, the policies of the 1990s and early 2000s entrenched many of the racial disparities evident in incarceration rates today.
Cumulatively, this body of scholarship underscores both the historical persistence and contemporary mechanics of racial incarceration disparities in Louisiana. By focusing on the state's most disparate parishes, this project synthesizes and contributes to the aforementioned research by tracing structural inequality to localized policy results and carceral practices.


### Methodology & Data
For this project, we used data from the Vera Institute of Justice’s Incarceration Trends dataset, which provides yearly jail population numbers by race at the county level. We looked specifically at data from Louisiana parishes between 1990 and 2010, which includes the years before and after the 1994 Crime Bill and also covers the period around Hurricane Katrina in 2005.

To measure racial disparity, we calculated a Black-to-white incarceration disparity ratio for each parish and year by dividing the Black incarceration rate by the white incarceration rate. We then found the average disparity ratio for each parish over the 20-year period and identified the parishes with the highest overall disparities — with Livingston Parish standing out the most.

After filtering the data to Louisiana only, we used line plots to explore how disparity ratios changed over time in individual parishes. We compared parishes with very different patterns — like Livingston (which had a huge spike around 2005) and St. Bernard (which stayed more stable). We also created a choropleth map to visualize how average disparity ratios varied across all parishes in the state.

To see whether major policy events made a difference, we compared disparities before and after 1994 using a boxplot and a t-test, and grouped the data by presidential administration to check for larger political trends. Then, to go a step further, we ran a regression analysis using Black and white incarceration rates as predictors to see how much they influenced the disparity ratio.

Throughout the analysis, we wrote explanations for each step.

### Findings & Analysis
After narrowing the data to Louisiana parishes from 1990 to 2010, we calculated the average Black-to-white incarceration disparity ratio for each parish. The results revealed major differences across the state. The five parishes with the highest average disparity ratios were Livingston, Cameron, Plaquemines, Lafourche, and St. Bernard, with some showing Black residents being incarcerated at 10 times the rate of white residents. These parishes became the focus of our deeper analysis.

We looked closely at Livingston Parish, which stood out for both the severity and timing of its disparity spike. In 2005, the disparity ratio in Livingston jumped to over 40, driven by a sharp increase in the Black jail population while the white jail population stayed almost unchanged. This wasn’t a statistical outlier — it reflected a real shift in incarceration patterns. The timing also lined up with Hurricane Katrina, which brought displacement, changes in local jail capacity, and possibly stricter enforcement in response to the crisis. This suggests that major events can drastically alter racial incarceration dynamics, especially in smaller or less-resourced areas.

In comparison, St. Bernard Parish also had a high average disparity ratio but showed a more stable trend. A small spike appeared in 2006, but the disparity didn’t reach Livingston’s levels and didn’t persist. This contrast highlights how local systems and responses — not just state or federal policy — shape racial outcomes. Even neighboring parishes may react differently to the same events based on leadership, resources, or historical patterns of policing.

To understand how these disparities were distributed geographically, we created a choropleth map showing average disparity ratios by parish. The map made it clear that racial disparities were not evenly spread. A few parishes showed relatively balanced incarceration rates, but others — like Livingston and Plaquemines — had extreme disparities. This supported our decision to focus on local variation, and underscored how place-based differences are central to understanding racial injustice in incarceration.

Next, we tested the impact of federal criminal justice policy, particularly the 1994 Crime Bill, using a pre- and post-policy comparison. A boxplot showed that after 1994, disparities were not only slightly higher but also more spread out. A t-test confirmed the difference was statistically significant, suggesting that the Crime Bill was associated with a measurable shift in racial incarceration disparities across Louisiana. This policy-level impact adds weight to the idea that national legislation can influence local systems, especially when paired with state-level implementation.

We also grouped data by presidential administration. The highest average disparities occurred under Clinton, aligning with the Crime Bill and the peak of “tough-on-crime” policy. Disparities began to decline under Bush Jr. and continued under Obama, reflecting a shift in national tone and possible changes in funding or enforcement focus. While federal administrations don’t directly manage jails, they shape the environment in which local systems operate.

Finally, we ran a regression analysis to explore which factors were most closely tied to high disparity ratios. Using Black and white incarceration rates as predictors showing higher Black incarceration rates were strongly associated with higher disparity ratios and higher white incarceration rates were associated with lower disparity ratios.

Disparities are largely driven by differences in how each group is policed, detained, and sentenced, not by random variation. The gap grows most when Black jail populations increase and white rates remain low — exactly what we saw in Livingston Parish.

Racial incarceration disparities in Louisiana’s most unequal parishes were shaped by a combination of federal policy shifts, major events like Katrina, and local-level practices.

```python
print(vera_df.head())
```

```python
#cleaning up vera dataset
vera_df = vera_df[(vera_df['year'] >= 1990) & (vera_df['year'] <= 2010)]
cols = [
    'year', 'county_name', 'state', 'fips',
    'black_jail_pop', 'white_jail_pop',
    'black_pop_15to64', 'white_pop_15to64'
]

vera_df = vera_df[cols]

vera_df = vera_df.dropna(subset=[
    'black_jail_pop', 'white_jail_pop',
    'black_pop_15to64', 'white_pop_15to64'
])

vera_df = vera_df[
    (vera_df['black_pop_15to64'] > 0) &
    (vera_df['white_pop_15to64'] > 0)
]

vera_df['black_rate'] = vera_df['black_jail_pop'] / vera_df['black_pop_15to64']
vera_df['white_rate'] = vera_df['white_jail_pop'] / vera_df['white_pop_15to64']

vera_df = vera_df[vera_df['white_rate'] > 0]

vera_df['disparity_ratio'] = vera_df['black_rate'] / vera_df['white_rate']

national_trends = vera_df.groupby('year')[['black_rate', 'white_rate', 'disparity_ratio']].mean().reset_index()
```

```python
louisiana_df = vera_df[vera_df['state'] == 'LA']
louisiana_df['county_name'].value_counts()
```

After filtering the dataset to only include rows where state == 'LA', we ran .value_counts() on the county_name column. This gives us a count of how many data entries exist per parish (Louisiana’s equivalent of counties) between 1990 and 2010.

Most parishes — like Acadia, Allen, Ascension, etc. — have 21 entries, which aligns with the 21-year range (1990–2010). This suggests strong, consistent coverage across time for those areas.

However, a few parishes — such as Madison (18 entries) and Plaquemines and Tensas (15 each) — have some missing years. This will be important to keep in mind when we later compare trends across parishes, especially if we’re doing any kind of time series analysis or calculating averages.

This consistency makes Louisiana a strong candidate for visualizations and deeper geographic breakdowns, since most areas are well represented over time.

```python
la_disparities = louisiana_df.groupby('county_name')['disparity_ratio'].mean().reset_index().sort_values(by='disparity_ratio', ascending=False)
la_disparities.head()
```

Here, we calculate the average racial disparity ratio for each Louisiana parish over the full 1990–2010 period. The disparity ratio measures how many times higher the Black incarceration rate is compared to the white rate in a given parish and year.

The five parishes with the highest average disparities are:
St. Bernard Parish	11.80
Livingston Parish	11.80
Plaquemines Parish	11.48
Lafourche Parish	11.17
Cameron Parish	10.99

These numbers suggest unequal incarceration patterns, where Black residents were, on average, jailed at 10–12 times the rate of white residents — far above national averages.

So this gives a meaningful starting point for examining why certain parishes have such extreme disparities and whether those trends intensified or improved over time, especially around the 1994 Crime Bill or state-level policy moments.

```python
#naif
top_parishes = la_disparities.head(5)['county_name'].tolist()
top_parishes_df = louisiana_df[louisiana_df['county_name'].isin(top_parishes)]

plt.figure(figsize=(12, 6))
sns.lineplot(data=top_parishes_df, x='year', y='disparity_ratio', hue='county_name', marker='o')
plt.title('Disparity Ratio Over Time in Top 5 Louisiana Parishes (1990–2010)')
plt.ylabel('Black-to-White Incarceration Disparity Ratio')
plt.xlabel('Year')
plt.grid(True)
plt.legend(title='Parish')
plt.tight_layout()
plt.show()
```

This line chart shows how the racial disparity in incarceration changed over time in the five parishes in Louisiana with the highest average disparities. Each line tracks how many times more likely Black residents were to be jailed compared to white residents from 1990 to 2010.

You can immediately see that the patterns are pretty different across parishes. For example, Livingston Parish has a huge spike around 2005 where the disparity ratio jumps above 45, which is extreme. Cameron Parish also shows a big spike during the early 2000s, while Plaquemines Parish had high but more stable disparities in the 90s before dropping. On the other hand, St. Bernard and Lafourche show more consistent trends, staying above 10 for much of the period.

This really shows how local these disparities are — it’s not just a statewide or national trend. Something specific might’ve happened in these counties: policy changes, shifts in policing, jail expansion, or even something like Hurricane Katrina affecting population numbers or jail records around 2005.

We’re starting to get a clearer picture, but now it makes sense to pick one of these parishes and look more closely at what’s driving those numbers.

```python
#naif
livingston_df = louisiana_df[louisiana_df['county_name'] == 'Livingston Parish']

plt.figure(figsize=(10,6))
sns.lineplot(data=livingston_df, x='year', y='black_rate', label='Black Incarceration Rate', marker='o')
sns.lineplot(data=livingston_df, x='year', y='white_rate', label='White Incarceration Rate', marker='o')
plt.title('Incarceration Rates by Race in Livingston Parish (1990–2010)')
plt.ylabel('Incarceration Rate (Jail Pop / Pop 15–64)')
plt.xlabel('Year')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

This plot breaks down the Black and white incarceration rates in Livingston Parish from 1990 to 2010. What it shows is that the spike in racial disparity wasn’t caused by some huge rise in white incarceration — it was almost entirely driven by a sudden jump in the Black incarceration rate around 2005.

The Black rate shoots up dramatically that year, while the white rate stays flat (and honestly, just hovers very low throughout the whole period). That explains why the disparity ratio exploded in that one year — it wasn’t about both groups being jailed more, it was about one group being jailed way more.

After 2005, the Black rate starts to drop gradually but still stays much higher than it was in the early 90s. That big spike raises the question of what happened in Livingston Parish around 2005.

This makes it worth exploring not just the ratios, but also the population sizes — because even small population changes could cause these numbers to swing hard.

```python
#naif
plt.figure(figsize=(10,6))
sns.lineplot(data=livingston_df, x='year', y='black_pop_15to64', label='Black Population (15–64)', marker='o')
sns.lineplot(data=livingston_df, x='year', y='white_pop_15to64', label='White Population (15–64)', marker='o')
plt.title('Black and White Population (15–64) in Livingston Parish (1990–2010)')
plt.ylabel('Population')
plt.xlabel('Year')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

This shows how the Black and white working-age populations (ages 15–64) changed in Livingston Parish between 1990 and 2010.

The white population steadily increases the entire time — going from around 44,000 in 1990 to nearly 80,000 by 2010. The Black population is much smaller, and while it also grows, the numbers stay low — from just over 2,000 in 1990 to around 4,000 by 2010.

The most important takeaway here is how small the Black population is. When a population is this small, even slight increases in the number of people incarcerated can cause the incarceration rate to spike — which is likely part of what caused that massive jump in the disparity ratio around 2005. A small denominator makes the math swing more dramatically.

So it’s possible that in 2005, a relatively small increase in the Black jail population had a huge effect on the rate — especially if the population estimate didn’t keep up with real-time changes

This kind of geographic variation supports the idea that local factors — not just statewide or federal policy — are driving a lot of the racial disparities we’re seeing.

To understand how polciy shifts may have shaped these trends, we want to look at incarceration disparities before and after the 1994 Crime Bill.

```python

louisiana_df['policy_period'] = louisiana_df['year'].apply(lambda x: 'Pre-1994' if x < 1994 else 'Post-1994')

# disparity ratios before vs after 1994
plt.figure(figsize=(8,5))
sns.boxplot(data=louisiana_df, x='policy_period', y='disparity_ratio')
plt.title('Disparity Ratios in Louisiana: Pre- vs Post-1994 Crime Bill')
plt.ylabel('Black-to-White Incarceration Disparity Ratio')
plt.grid(True)
plt.tight_layout()
plt.show()

# check for statistical difference
from scipy.stats import ttest_ind

pre_1994 = louisiana_df[louisiana_df['year'] < 1994]['disparity_ratio']
post_1994 = louisiana_df[louisiana_df['year'] >= 1994]['disparity_ratio']

t_stat, p_val = ttest_ind(pre_1994, post_1994, equal_var=False)

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_val:.5f}")
```

This boxplot compares incarceration disparity ratios in Louisiana before and after the 1994 Crime Bill. While the median values look similar, the spread is wider after 1994, and there are more extreme outliers. That means racial disparities in incarceration became more uneven after the bill passed.

The t-test backs this up — the difference is statistically significant (T = 3.153, p = 0.00176). So something really did change after 1994, even if the reasons aren’t fully clear just from this chart.

To keep exploring how national policy might have shaped these trends, we want to look at average disparity ratios under each presidential administration. This could helps show whether certain time periods were linked to higher or lower levels of racial disparity, and adds more context to how different admins affect this.

```python
#arsalan
# each year to presidential administration
def get_admin_period(year):
    if 1989 <= year <= 1992:
        return 'Bush Sr.'
    elif 1993 <= year <= 2000:
        return 'Clinton'
    elif 2001 <= year <= 2008:
        return 'Bush Jr.'
    elif 2009 <= year <= 2010:
        return 'Obama'
    else:
        return 'Other'

louisiana_df['admin_period'] = louisiana_df['year'].apply(get_admin_period)

admin_avg = louisiana_df.groupby('admin_period')['disparity_ratio'].mean().reset_index()

admin_order = ['Bush Sr.', 'Clinton', 'Bush Jr.', 'Obama']
admin_avg['admin_period'] = pd.Categorical(admin_avg['admin_period'], categories=admin_order, ordered=True)
admin_avg = admin_avg.sort_values('admin_period')

plt.figure(figsize=(8,5))
sns.barplot(data=admin_avg, x='admin_period', y='disparity_ratio', palette='Blues_d')
plt.title('Average Incarceration Disparity by Presidential Administration (Louisiana)')
plt.ylabel('Disparity Ratio')
plt.xlabel('Administration')
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```

This chart shows average incarceration disparity ratios in Louisiana during each presidential administration from 1990 to 2010. The disparities were highest during the Clinton administration, peaking at nearly 7 to 1, and gradually declined during Bush Jr. and Obama.

This trend is important when considering Louisiana’s most unequal parishes — like Livingston and St. Bernard — because many of their highest disparity years happened during or shortly after the Clinton era, especially around the mid-1990s to early 2000s. That aligns with the 1994 Crime Bill, which expanded jail funding, promoted harsher sentencing, and pushed states toward more punitive policies. Even though the bill was federal, it directly influenced the jail infrastructure and practices in parishes across Louisiana.

By the time of the Bush and Obama years, racial disparities began to level off or decline in many of these same parishes. This suggests that national policy climate did have a noticeable impact, especially when paired with state and local responses to federal funding and priorities.

Also, to better understand what’s actually driving these disparity numbers at a more technical level we want to run a simple regression

```python
import statsmodels.api as sm

#evan
regression_df = louisiana_df[['disparity_ratio', 'black_rate', 'white_rate']].dropna()

#  disparity_ratio using black and white incarceration rates
X = regression_df[['black_rate', 'white_rate']]
y = regression_df['disparity_ratio']
X = sm.add_constant(X)
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())

plt.figure(figsize=(10, 6))
plt.scatter(regression_df['black_rate'],
            regression_df['disparity_ratio'],
            alpha=0.3)
plt.xlabel('Black Incarceration Rate')
plt.ylabel('Disparity Ratio')
plt.title('How Black and White Incarceration Rates Impact Disparity Ratios')
plt.grid(True)

x_range = np.linspace(regression_df['black_rate'].min(), regression_df['black_rate'].max(), 100)
y_pred = model.params[0] + model.params[1] * x_range + model.params[2] * regression_df['white_rate'].mean()
plt.plot(x_range, y_pred, 'r--', linewidth=2, label=f'Trend at avg white rate: {regression_df["white_rate"].mean():.3f}')
plt.legend()

plt.tight_layout()
plt.show()
```

We ran a regression using Black and white incarceration rates as predictors.

The results show that both predictors are statistically significant:

The Black incarceration rate has a positive coefficient (+22.72), meaning that higher Black jail rates are strongly associated with higher disparity ratios.

The white incarceration rate has a negative coefficient (–50.86), suggesting that as white incarceration increases, the disparity ratio decreases.


This fits what we saw earlier in places like Livingston Parish, where a high Black rate and low white rate led to extreme disparities. The regression confirms that these two rates play a big role in shaping the racial gap.
