# **1. Introduction**

In recent years, scholars and policymakers have become increasingly focused on both the structural and psychological barriers that encourage and inhibit political participation in the United States. A growing body of research has examined voting through the lens of a cost-benefit analysis, emphasizing how factors such as travel distance, time constraints, and opportunity costs all potentially inhibit would-be voters. While substantial literature has investigated how the distance between a voter's home and polling locations impact voter turnout, there has been far less attention on how commute distances to work, specifically via public transportation, impact an individual's perceived or actual ability and cost associated with voting.

This gap in research is particularly important in dense urban settings such as New York City, where nearly half of all households rely on public transit and where commute times can vary drastically for individuals living across the five boroughs.

In this paper, we examine how commuting distance to work, particularly by public transit, affects voter turnout in New York City across the 2016 and 2020 presidential elections. We test the commuter's strain hypothesis in a highly public-transit-relient context, exploring whether long public transit commutes act as a hidden cost that depresses voter turnout even when physical polling access is relatively high.

#**1.1 Related Work**




There currently exists significant literature on the “costs” associated with voting and how those costs impact voter turnout. Much of the literature focuses on cost in terms of distance from polling locations. However, to our knowledge, prior research does not take into account how individuals’ commute distance to work using public transit impacts voter turnout from a perspective of adding a further “cost” to voting.

The majority of the research available on the impacts of distance to polling location and the “costs” associated with voting focus their analysis exclusively from a national lens or on states with significantly lower public-transit usage rates than New York City. The idea of the “commuter’s strain” hypothesis was introduced which emphasizes time spent in transit as a more psychologically burdensome activity than the workday itself [4]. As a result of this, time spent working has no effect on voter turnout while commuting time significantly decreased political participation [4]. However, these effects failed to be considered in a highly metropolitan setting such as New York City where 48% of households rely on public transportation for work as compared to the national 4% [5]. It was found in Brazil during the 2022 national election that implementing fare-free public transit during the election had no significant increase in voter turnout despite a measurable rise in ridership on election days [9]. This suggests that while physical access was improved, deeper psychological and logistical barriers such as time constraints, informational deficits, or stress remained unaddressed. In contexts like New York City, where public transportation is widely available, this evidence complicates the assumption that access alone ensures participation. It reinforces the argument that voting deterrents like the psychological toll of commuting, or “commuter’s strain”, may outweigh any structural improvements to access [9]. Additionally, some attribute diminished voter turnout to the decrease in voters’ perceived relevance of their ballot [11]. We hope to explore the relationship between commuting distance to work and voter turnout specifically in an area with an abundance of public transit resources to understand how a high reliance on public transit affects political participation. Furthermore, in accordance with the “commuter’s strain” hypothesis, we aim to understand how the psychological impacts of commuting might differ for commuting via car, as examined in previous research, versus public transit to then understand how those potentially differing psychological strains impact voter turnout.

There is a wide-spread general consensus across the literature that increasing distance to polling locations decreases voter turnout [1, 2, 6, 7, 10]. This effect is nearly twice as large for voters without reliable vehicle access, demonstrating that even marginal travel burdens can meaningfully suppress turnout [7]. In fact, it was found that a one-mile increase in road-network distance depresses habitual voter turnout by roughly 1.5 percent, which is especially alarming as it is roughly twice the average effect found in non-habitual voters [10]. This shows how even marginal increases in the burdens of travel can dissuade even the most motivated voters.

Individuals at polls that are about the same distance from their residence as other amenities had a higher voter turnout than those voters whose distance to the polls was twice that of the distance to other amenities [1]. This brings up an interesting idea of how voters gauge distance in terms of the comparison between distance to voting sites and distance to places they frequently visit. Given that one’s workplace would be considered a location they likely frequently visit, the relationship between distance to work and distance to polling location can be explored in terms of how individuals with longer or shorter commutes might view a longer or shorter distance to a polling location more or less favorably.

The perception of distance in terms of polling location in comparison to distance to other frequently visited locations ties into the overall concept introduced by studies of voting as a cost-benefit calculation. Researchers argue that a voter calculates their benefit by multiplying the perceived benefit that the voter will receive from their desired political outcome by the probability that the voter will cast the marginally important vote [3]. From this value, the voter subtracts their cost to decide whether to vote or not. Voters with longer distances to polling locations in turn have high travel costs, inevitably reducing the likelihood that their cost-benefit analysis will point them to vote [3]. Furthermore, travel time to precinct sites also imposes an opportunity cost given that it takes away time from work or leisure time that could be spent on other activities [3]. In terms of commute distance to work, voters are then inclined to consider how their commute to work and potential disturbances to commuting will be impacted by their decision to vote or not.

A prior study also highlighted how the effects of distance to polling locations is much larger in areas with a heavy reliance on public transportation in terms of decreasing the likelihood of voting [6]. While this study does not focus on public transportation from a commuting to work perspective, it does shed light on the fact that the distances associated with commuting to work could potentially have an impact on the willingness of individuals to use public transportation (assuming that is their primary mode of transportation) to travel again to a polling location (presumably either before or after the workday). For North Carolina voters, it was found that being reassigned one mile farther away reduces in‐person Election Day voting by about 2.5 percentage points. Although much of this “lost” turnout reappears via early ballots, the substitution effect is substantially weaker among those whose new sites lie outside public‐transit corridors. This suggests that in a transit-rich city like New York, longer work commutes may compound and worsen precinct‐change burdens in ways early-vote options cannot fully offset [8].

We aim to explore how this contributes to the overall “commuter’s strain” hypothesis mentioned above as well as to the cost-benefit analysis associated with political participation in terms of examining the effects of the potentially added difficulty in getting to a polling location while commuting via public transportation for work in New York City.

We began by pulling tract-level data from the U.S. Census Bureau's 2016 American Community Survey (ACS) for all five boroughs of New York City: Bronx, Manhattan, Queens, Brooklyn, and Staten Island. Specifically, we collected demographic information pertinent to our analysis such as mean commute time and population eligible to vote. We also collected commuting data that captures how many workers in each tract commute more than 45 minutes. After conducting a sensitivity analysis, we found 45 minutes to be an effective distinguisher for what would classify as a "long" commute.

For each NYC county, the script makes two API requests: one for high-level demographic statistics and another for commute duration breakdowns. The results are stored in lists, then concatenated into two citywide datasets. We then clean the data by converting all numerical values from strings to proper numeric types and compute a new variable: the share of long commuters (those traveling 45 minutes or more). Finally, the code constructs a standardized tract identifier (GEOID) and merges the two datasets into a single DataFrame (acs_merged), which is used in later analysis to explore the relationship between commuting patterns and voter turnout.


```python
import pandas as pd
import requests

# List of NYC county FIPS codes: Bronx(005), Brooklyn/Kings(047), Manhattan/New York(061), Queens(081), Staten Island/Richmond(085)
counties = ['005','047','061','081','085']

# ACS 5-year 2012-2016 data: mean travel time (DP03_0025E) and population 18+ (DP05_0018E) from profile,
# and commute time distribution from detailed table B08134 (total workers, and workers with 45-59 and 60+ min commutes)
acs_profile_list = []
acs_commute_list = []
for county in counties:
    # ACS profile API for mean travel time and 18+ population
    profile_url = "https://api.census.gov/data/2016/acs/acs5/profile"
    params = {
        "get": "DP03_0025E,DP05_0018E,NAME",
        #"get": "DP03_0025E,B29001_001E,NAME",
        "for": f"tract:*",
        "in": f"state:36 county:{county}"
        # "key": "YOUR_CENSUS_API_KEY"  # (Include your API key if you have one)
    }
    resp = requests.get(profile_url, params=params)
    data = resp.json()
    df_profile = pd.DataFrame(data[1:], columns=data[0])
    acs_profile_list.append(df_profile)
    # ACS detailed table B08134 for commute time distribution
    detail_url = "https://api.census.gov/data/2016/acs/acs5"
    params = {
        "get": "B08134_001E,B08134_009E,B08134_010E",  # total commuters, commuters 45-59 min, commuters 60+ min
        "for": f"tract:*",
        "in": f"state:36 county:{county}"
    }
    resp = requests.get(detail_url, params=params)
    data = resp.json()
    df_commute = pd.DataFrame(data[1:], columns=data[0])
    acs_commute_list.append(df_commute)

# Concatenate data for all counties
acs_profile = pd.concat(acs_profile_list, ignore_index=True)
acs_commute = pd.concat(acs_commute_list, ignore_index=True)

# Convert numeric fields from string to proper numeric types
acs_profile['DP03_0025E'] = pd.to_numeric(acs_profile['DP03_0025E'])   # mean travel time (minutes)
acs_profile['DP05_0018E'] = pd.to_numeric(acs_profile['DP05_0018E'])   # population 18+ (for turnout rate denominator)
acs_commute['B08134_001E'] = pd.to_numeric(acs_commute['B08134_001E']) # total workers (did not work at home)
acs_commute['B08134_009E'] = pd.to_numeric(acs_commute['B08134_009E']) # commuters 45-59 min
acs_commute['B08134_010E'] = pd.to_numeric(acs_commute['B08134_010E']) # commuters 60+ min

# Compute share of long commutes (>45 min)
acs_commute['long_commute_share'] = (acs_commute['B08134_009E'] + acs_commute['B08134_010E']) / acs_commute['B08134_001E']

# Merge the ACS profile and commute data on tract identifiers
# Create a tract GEOID (state+county+tract code) for merging
acs_profile['GEOID'] = acs_profile['state'] + acs_profile['county'] + acs_profile['tract']
acs_commute['GEOID'] = acs_commute['state'] + acs_commute['county'] + acs_commute['tract']
acs_merged = pd.merge(acs_profile, acs_commute[['GEOID','long_commute_share']], on='GEOID')
acs_merged.head(6)
```

From there, we load and prepare both election data and geographic boundaries so they can be linked spatially. We filter and aggregate 2016 NYC presidential vote data at the election district level and align it with official NYC shapefiles by reprojecting and renaming fields. This is a critical preprocessing step that allows for merging vote counts with geographic units (census tracts) for turnout analysis and mapping.

```python
import pandas as pd
import geopandas as gpd

# 1. Load and aggregate ED‐level vote totals
ed_2016_results = pd.read_csv("NYC_2016_PresEDLevel.csv")
ed_2016_results['Tally'] = pd.to_numeric(ed_2016_results['Tally'], errors='coerce')

ballot_categories = [
    "Public Counter", "Manually Counted Emergency",
    "Absentee / Military", "Federal", "Special Presidential", "Affidavit"
]
ed_2016_totals = (
    ed_2016_results[ed_2016_results['Unit Name'].isin(ballot_categories)]
      .groupby(['County','AD','ED'], as_index=False)['Tally']
      .sum()
      .rename(columns={'Tally':'ballots'})
)
ed_2016_totals['ElectDist'] = ed_2016_totals['AD'] * 1000 + ed_2016_totals['ED']

# 2. Load shapefiles
ed_shp = gpd.read_file("ed_data/Election_Districts.shp")
tracts_shp = gpd.read_file("tl_2016_36_tract.shp")

tracts_shp = tracts_shp[tracts_shp['COUNTYFP'].isin(['005','047','061','081','085'])]

# **Reproject ED layer to tract CRS**
ed_shp = ed_shp.to_crs(tracts_shp.crs)
ed_shp = ed_shp.rename(columns={'Election_D': 'ElectDist'})

ed_shp.head(3)
```

Afterwards, we perform a spatial mapping between Election Districts (EDs) and Census Tracts by assigning each election district to the census tract that contains its centroid. This mapping is essential for aggregating vote totals (originally reported at the ED level) to tract-level geography, which aligns with ACS demographic and commute data.

```python
# Reproject ED shapefile to match tract CRS before centroid
ed_shp = ed_shp.to_crs(tracts_shp.crs)

# If tract CRS is still geographic, reproject both to a projected CRS
# EPSG:2263 is good for NYC
projected_crs = "EPSG:2263"
ed_shp = ed_shp.to_crs(projected_crs)
tracts_shp = tracts_shp.to_crs(projected_crs)

# 3. Build centroids and spatial-join to map ED → tract
ed_centroids = ed_shp[['ElectDist', 'geometry']].copy()
ed_centroids['geometry'] = ed_centroids.geometry.centroid

tracts_geom = tracts_shp[['GEOID', 'geometry']].copy()

#This is joining the ed_centroid to tract by 'geometry' value (the tract has to contain a ed centroid to be kept)
#Only the ElectionDistrict value and Geoid (tract) value are kept
ed_to_tract = gpd.sjoin(
    ed_centroids,
    tracts_geom,
    how='inner',
    predicate='within'
)[['ElectDist', 'GEOID']]
ed_to_tract['ElectDist'] = ed_to_tract['ElectDist'].astype(int)

print("ED→tract mappings:", len(ed_to_tract),
      "unique EDs:", ed_to_tract['ElectDist'].nunique())

#ed_to_tract.head(10)
```

Next, we merge together the Election District ballot totals with the Tract identifier.

```python
# 4. Merge vote totals into tract mapping and aggregate
ed_2016_with_tract = pd.merge(
    ed_2016_totals,
    ed_to_tract,
    on='ElectDist',
    how='inner'
)
print("Rows after merging ED votes with tracts:", len(ed_2016_with_tract))

#ed_2016_with_tract.head()
```

```python
tract_2016_turnout = (
    ed_2016_with_tract
      .groupby('GEOID', as_index=False)['ballots']
      .sum()
)
```

Finally, we calculate our turnate rate by using the ballots case and the voting age population (18+)

```python
# 5. Calculate turnout rate (%) using ACS adult population
tract_2016_turnout = pd.merge(
    tract_2016_turnout,
    acs_profile[['GEOID','DP05_0018E']],
    on='GEOID',
    how='inner'
)
tract_2016_turnout['turnout_rate_pct'] = (
    tract_2016_turnout['ballots'] / tract_2016_turnout['DP05_0018E']
) * 100

#NOTE: DP05_0018E is most likely a voting age population estimate
#tract_2016_turnout.head(3)
```

```python
tract_2016_data = tract_2016_turnout
```

We merge the tract data with the ACS data containing the mean commute time for each tract. This is the relationship we want to explore more – while there are other ways to quantify commute difficulty, we chose time over distance because we felt that it was a more accurate measure of how hard a commute was, especially in the city where public transportation is so popular.


```python
# Merge ACS commute data with the turnout & distance data
final_df = pd.merge(tract_2016_data, acs_merged[['GEOID','DP03_0025E','long_commute_share']], on='GEOID')
# Rename columns for clarity
final_df.rename(columns={
    'DP03_0025E': 'mean_commute_time',
    'long_commute_share': 'pct_long_commute',
    'ballots': 'votes_2016'
}, inplace=True)
final_df = final_df[['GEOID','mean_commute_time','pct_long_commute','votes_2016','turnout_rate_pct']]

"""
Ensure that both the mean commute time is a positive value AND
turnout rate pct is betwen [0,100]
"""
df_clean = final_df[final_df['mean_commute_time'] >= 0]
df_clean = df_clean[df_clean['turnout_rate_pct'] <= 100]
df_clean = df_clean[df_clean['turnout_rate_pct'] >= 0]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.regplot(
    data=df_clean,
    x='mean_commute_time',
    y='turnout_rate_pct',
    scatter_kws={'alpha': 0.5},
    line_kws={'color': 'red'}
)
plt.xlabel('Mean Commute Time (minutes)')
plt.ylabel('Voter Turnout Rate (%)')
plt.title('Commute Time vs. Voter Turnout Rate by Census Tract (NYC, 2016) - Arjun')
plt.tight_layout()
plt.show()
```

This is a scatterplot of voter turnout rate vs mean commute time, where each dot represents a tract. There is a large cluster of dots and there does not appear to be any obvious relationship between the two variables. Given that there was no obvious visual relationship, we added in a trend line to understand any potential connections between the variables.

The regression line highlights a list of tracts that could potentially point to some form of correlation between mean commute team and voter turnout rate so we then decided to plot each of these tracts that exist along the regression line on a map to explore the nature of the regions and determine any potential similarities.

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import geopandas as gpd

# STEP 1: Fit regression line on commute vs turnout
X = df_clean['mean_commute_time']
y = df_clean['turnout_rate_pct']
X_const = sm.add_constant(X)
model = sm.OLS(y, X_const).fit()
df_clean['predicted_turnout'] = model.predict(X_const)
df_clean['residual'] = np.abs(df_clean['turnout_rate_pct'] - df_clean['predicted_turnout'])

# STEP 2: Flag tracts close to trend line
threshold = 1  # within 1 percentage points of the trend
highlight_geoids = df_clean[df_clean['residual'] <= threshold]['GEOID']

# STEP 3: Load NYC-only shapefile just like the WFH plot
gdf_2016_tract = gpd.read_file("tl_2016_36_tract.shp")
nyc_fips = ['005', '047', '061', '081', '085']
nyc_tracts = gdf_2016_tract[gdf_2016_tract['COUNTYFP'].isin(nyc_fips)].copy()

# STEP 4: Merge shapefile with highlight flag
nyc_tracts['highlight'] = nyc_tracts['GEOID'].isin(highlight_geoids)

# STEP 5: Plot
fig, ax = plt.subplots(1, 1, figsize=(12, 12))
nyc_tracts.plot(ax=ax, color='lightgray', edgecolor='white')  # all tracts
nyc_tracts[nyc_tracts['highlight']].plot(ax=ax, color='red', edgecolor='black')  # trend-aligned tracts
ax.set_title('NYC Tracts along Commute–Turnout Trend (2016) - Sachin', fontsize=16)
ax.axis('off')
plt.tight_layout()
plt.show()
```

This map highlights tracts that are within 1 percentage point of the highlighted trend line. These tracts were highlighted to gain a better understanding of any relationship between them in attempts to further understand the trend line that was visualized earlier. From this map, we determined that there existed few similarities in the nature of these tracts from an income, work-from-home rate, and demographic perspective given how they are spread out across various boroughs. This led us to believe that the tracts pinpointed by the regression line were likely as a result of randomness rather than concrete connection.

```python
ed_2020_results = pd.DataFrame(parsed_rows, columns=col_names)

#Convert AD and ED to integers
ed_2020_results["AD"] = pd.to_numeric(ed_2020_results["AD"], errors="coerce").astype("Int64")
ed_2020_results["ED"] = pd.to_numeric(ed_2020_results["ED"], errors="coerce").astype("Int64")

ed_2020_results['Tally'] = pd.to_numeric(ed_2020_results['Tally'], errors='coerce')

ballot_categories = [
    "Public Counter", "Manually Counted Emergency",
    "Absentee / Military", "Federal", "Special Presidential", "Affidavit"
]
#sum across all ballot categories for each election district
ed_2020_totals = (
    ed_2020_results[ed_2020_results['Unit Name'].isin(ballot_categories)]
      .groupby(['County','AD','ED'], as_index=False)['Tally']
      .sum()
      .rename(columns={'Tally':'ballots'})
)

#same process to get the election district identifies from the AD and the ED
ed_2020_totals['ElectDist'] = ed_2020_totals['AD'] * 1000 + ed_2020_totals['ED']

# Merge vote totals into tract mapping and aggregate
ed_2020_with_tract = pd.merge(
    ed_2020_totals,
    ed_to_tract,
    on='ElectDist',
    how='inner'
)

# Sum across tracts with the same geoid
tract_2020_turnout = (
    ed_2020_with_tract
      .groupby('GEOID', as_index=False)['ballots']
      .sum()
)

# Calculate turnout rate (%) using ACS adult population
tract_2020_rate_merged = pd.merge(
    tract_2020_turnout,
    acs_2020_profile[['GEOID','Total']],
    on='GEOID',
    how='inner'
)
tract_2020_rate_merged['turnout_rate_pct'] = (
    tract_2020_rate_merged['ballots'] / tract_2020_rate_merged['Total']
) * 100

#tract_turnout_2020_data = pd.merge(tract_2016_turnout, tract_dist, on='GEOID')
tract_turnout_2020_data = tract_2020_rate_merged

# Merge ACS commute data with the turnout & distance data
final_2020_df = pd.merge(tract_turnout_2020_data, acs_2020_merged[['GEOID','DP03_0025E','long_commute_share']], on='GEOID')
# Rename columns for clarity
final_2020_df.rename(columns={
    'DP03_0025E': 'mean_commute_time',
    'long_commute_share': 'pct_long_commute',
    'ballots': 'votes_2020'
}, inplace=True)
final_2020_df = final_2020_df[['GEOID','mean_commute_time','pct_long_commute','votes_2020','turnout_rate_pct']]
final_2020_df.head(5)


# Perform the same cleaning
df_2020_clean = final_2020_df[final_2020_df['mean_commute_time'] >= 0]
df_2020_clean = df_2020_clean[df_2020_clean['turnout_rate_pct'] <= 100]
df_2020_clean = df_2020_clean[df_2020_clean['turnout_rate_pct'] >= 0]

df_2020_clean.head(15)
```

Next, based on a review of prior literature, there is a strong consensus that factors such as the "commuter's strain" hypothesis and the cost-benefit analysis associated with political participation impact low-income indivdidual more severely than others. Given that, we wanted to isolate low income tracts in New York City to understand whether that was a factor in the decision to the vote based again on commute to work time. Here, we investigated low-income tracts and indiduals that were considered to have a "long" commute based our sensitivity analysis and distinction described above.

```python
# 1. Download tract-level income data (ACS 2016, median household income)
income_url = "https://api.census.gov/data/2016/acs/acs5"
income_data = []
for county in counties:
    params = {
        "get": "B19013_001E,NAME",
        "for": "tract:*",
        "in": f"state:36 county:{county}"
    }
    resp = requests.get(income_url, params=params)
    data = resp.json()
    df_income = pd.DataFrame(data[1:], columns=data[0])
    income_data.append(df_income)

# Combine into single DataFrame
acs_income = pd.concat(income_data, ignore_index=True)

# Convert and create GEOID
acs_income["B19013_001E"] = pd.to_numeric(acs_income["B19013_001E"])
acs_income["GEOID"] = acs_income["state"] + acs_income["county"] + acs_income["tract"]
```

```python
# 1. Ensure consistent GEOID across datasets
#tract_2016_turnout["GEOID"] = tract_2016_turnout["GEOID"]

# 2. Merge datasets on 'geoid'
merged = tract_2016_turnout.merge(
    acs_commute, on="GEOID", how="inner"
).merge(
    acs_income, on="GEOID", how="inner"
)

# 3. Filter to low-income tracts (median household income < $40,000)
low_income = merged[merged["B19013_001E"] < 40000].copy()

# 4. Calculate long commute share if not already present
if "long_commute_share" not in low_income.columns:
    low_income["long_commute"] = low_income["B08134_009E"] + low_income["B08134_010E"]
    low_income["long_commute_share"] = low_income["long_commute"] / low_income["B08134_001E"]

# 5. Clean out unrealistic turnout rates (>100% or <=0%)
low_income_clean = low_income[
    (low_income["turnout_rate_pct"] > 0) &
    (low_income["turnout_rate_pct"] <= 100)
].copy()

# 6. Plot cleaned data
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(
    low_income_clean["long_commute_share"],
    low_income_clean["turnout_rate_pct"],
    alpha=0.6
)
plt.xlabel("Share of Long Commutes (45+ min)")
plt.ylabel("Voter Turnout Rate (%)")
plt.title("NYC 2016 Turnout vs Commute Length in Low-Income Tracts (2016) - Nikhil")
plt.grid(True)
plt.tight_layout()
plt.show()
```

The scatter suggests a mild downward trend: tracts with higher shares of 45+ minute commutes tend to report lower turnout rates. When fewer than about 20% of commuters face long travel, turnout is quite variable—ranging from under 20% to nearly 100%. In contrast, tracts where over half the workforce endures extended commutes seldom exceed roughly 60% turnout. The concentration around 30–55% long commutes and 20–50% turnout hints that moderate travel burdens coincide with mid-level engagement. Overall, these patterns may indicate that longer commutes are associated with somewhat reduced voter participation in low-income neighborhoods.

# **4. Results**

An analysis of the 2016 data shows that mean commute time by public transit bears virtually no relationship to voter turnout at the census-tract level. As we can see in our scatter plot of average commute time versus turnout percentage, there is no clear trend among the data points. The dots are concentrated in the middle of the graph and the added regression line is almost flat.

Complementary visualizations such as commute time versus median household income and a choropleth of work-from-home rates reinforce that commute burden alone does not explain variation in turnout across NYC tracts.

Repeating the analysis for the 2020 presidential election, during which many New Yorkers temporarily reduced their commutes, yields the same null result: no discernible correlation between average commute time and turnout. A focused look at low-income tracts (where we assumed commuter’s strain to be hit the hardest considering the hardships of the pandemic) also fails to uncover a negative relationship between the share of long (>45 min) commutes and turnout rates. Overall, these findings suggest that in a public transit-centric metropolis like New York City, the psychological or temporal cost of commuting (as described by the "commuter's strain" hypothesis and cost-benefit analysis of voting) via public transit does not by itself depress electoral participation.
